```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("彭瑞蘭.jpg")

sr = cv2.dnn_superres.DnnSuperResImpl_create()
sr.readModel("LapSRN_x8.pb")
sr.setModel("lapsrn", 8)

result = sr.upsample(img)

resized = cv2.resize(img, dsize=None, fx=8, fy=8)

plt.figure(figsize=(12,8))
plt.subplot(1,3,1)

# Original image
plt.imshow(img[:,:,::-1])
plt.subplot(1,3,2)

# SR upscaled
plt.imshow(result[:,:,::-1])
plt.subplot(1,3,3)

# OpenCV upscaled
plt.imshow(resized[:,:,::-1])
plt.show()
```

```python
cv2.imwrite("彭瑞蘭_sr.jpg", result)

files.download("彭瑞蘭_sr.jpg")
```
